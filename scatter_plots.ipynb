```python
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../metric_csv_files/MRI_GuysT1__20240309-215304_BS32_EP10IXI030-Guys-0708-T1050pro.csv')

images_list = ['VXM_MSE1', 'VXM_NCC', 'VXM_Simonovosky', 'VXM_DeepSIM', 'VXM_SIAM_PRO', 'ANTS']
titles = ['$\mathrm{VXM_{mse}}$', '$\mathrm{VXM_{ncc}}$', '$\mathrm{VXM_{Cheng}}$',
          '$\mathrm{VXM_{DeepSim}}$', '$\mathrm{VXM_{SiamRegQC}}$', '$\mathrm{ANTsPy}$']
metric_list  = ['MSE', 'SSIM']
fig, axs = plt.subplots(2, 6, figsize=(25, 10))
plt.subplots_adjust(hspace=0.35, wspace=0.25)
for i, metric in enumerate(metric_list):
    x_axis = np.array(df[f'Deformed_{metric}'])
    for j, (title, image_key) in enumerate(zip(titles, images_list)):
        y_axis = np.array(df[f'{image_key}_{metric}'])
        axs[i, j].scatter(x_axis, y_axis, s=25)
        axs[i, j].plot(x_axis, x_axis, '--r', alpha=0.6)
        axs[i,j].ticklabel_format(axis='y', style='sci', scilimits=(0,0))
        # axs[i,j].ticklabel_format(axis='x', style='sci', scilimits=(0,0))
        axs[i, j].tick_params(labelsize=20)
        axs[i,j].set_xlabel(f'initial {metric}\n(Before Registration)', fontsize=18)

for j, title in enumerate(titles):
    axs[0,j].set_title(title, fontsize=25)
    # axs[0,j].set_ylim(0.0, 0.08)
for i, metric in enumerate(metric_list):
    axs[i,0].set_ylabel(f'{metric}\n(After Registration)', fontsize=20)
            
# plt.savefig('scatter_plots.pdf', format='pdf', bbox_inches='tight')
```
